# **Задача 1**

## Исходные данные

$$
X(t)=3U\sin t-e^{3t}V+\cos t,
$$

где
$$
U\sim E(0.25), \quad V\sim R(2,4), U \text{ и } V \text{ некоррелированы.}
$$

---

## 1. Математическое ожидание
$$
m_X(t)=M[X(t)]=3\sin tM[U]-e^{3t}M[V]+\cos t.
$$

Моменты случайных величин:
$$
M[U]=\frac{1}{0.25}=4,
$$
$$
M[V]=\frac{2+4}{2}=3.
$$

Итого:
$$
m_X(t)=12\sin t-3e^{3t}+\cos t.
$$

---

## 2. Дисперсия

По определению:
$$
D_X(t)=M\Big[(X(t)-m_X(t))^2\Big].
$$

Получаем:
$$
X(t)-m_X(t)=3\sin t(U-4)-e^{3t}(V-3).
$$

Возведем в квадрат:
$$
(X(t)-m_X(t))^2=
9\sin^2t(U-4)^2
+e^{6t}(V-3)^2
-6e^{3t}\sin t(U-4)(V-3).
$$

Так как:
$$
M[(U-4)(V-3)]=0,
$$

Получаем:
$$
D_X(t)=9\sin^2tD_U+e^{6t}D_V.
$$

Дисперсии:
$$
D_U=\frac{1}{0.25^2}=16,
$$
$$
D_V=\frac{(4-2)^2}{12}=\frac{1}{3}.
$$

В итоге:
$$
D_X(t)=144\sin^2t+\frac{1}{3}e^{6t}.
$$

# **Задача 2**

## Исходные данные

$$
Z(t)=X(t)e^{-3t}-Y(t)\sin t-t,
$$

$$
K_X(t_1,t_2)=1+\cos(t_2-t_1),
$$
$$
K_Y(t_1,t_2)=\sin t_1\sin t_2,
$$

$$
X(t),Y(t) \text{ некоррелированы.}
$$

---

## 1. Математическое ожидание
Поскольку задан вид корреляционной функции и не задано мат. ожидание для каждого из процессов X и Y, полагаем, что процессы центрированы, т.е
$$
m_X(t)=0
$$
$$
m_Y(t)=0
$$

Тогда:
$$
m_Z(t)=-t.
$$

---

## 2. Дисперсия

По определению:
$$
D_Z(t)=M[(Z(t)-m_Z(t))^2].
$$

Так как:
$$
Z(t)-m_Z(t)=X(t)e^{-3t}-Y(t)\sin t,
$$

квадрат:
$$
(Z(t)-m_Z(t))^2=
e^{-6t}X^2(t)
+\sin^2tY^2(t)
-2e^{-3t}\sin tX(t)Y(t).
$$

Так как:
$$
M[X(t)Y(t)]=0,
$$

получаем:
$$
D_Z(t)=e^{-6t}K_X(t,t)+\sin^2tK_Y(t,t).
$$

Подстановка:
$$
K_X(t,t)=2,
$$
$$
K_Y(t,t)=\sin^2t.
$$

Итог:
$$
D_Z(t)=2e^{-6t}+\sin^4t.
$$

# **Задача 6**

## Исходные данные

$$
X(t)=f(t)U, \quad f(t)=e^{-3t},
$$
$$
Z(t)=\int_0^t X(s)ds,
$$
$$
U\sim P(5).
$$

---

## 1. Математическое ожидание процесса (X(t))

По определению:
$$
m_X(t)=M[X(t)]=f(t)M[U].
$$

Для распределения Пуассона:
$$
M[U]=5.
$$

Следовательно:
$$
m_X(t)=5e^{-3t}.
$$

---

## 2. Математическое ожидание процесса (Z(t))

$$
m_Z(t)=M\left[\int_0^t X(s)ds\right]
=\int_0^t m_X(s)ds.
$$

Получаем:
$$
m_Z(t)=5\int_0^t e^{-3s}ds.
$$

---

## 3. Корреляционная функция процесса (X(t))

По определению:
$$
K_X(t_1,t_2)=M[X(t_1)X(t_2)].
$$

Подставляем:
$$
K_X(t_1,t_2)=f(t_1)f(t_2)M[U^2].
$$

Для Пуассона:
$$
M[U^2]=D_U+M^2[U]=5+25=30.
$$

Следовательно:
$$
K_X(t_1,t_2)=30e^{-3(t_1+t_2)}.
$$

---

## 4. Корреляционная функция процесса (Z(t))

По определению:
$$
K_Z(t_1,t_2)=M[Z(t_1)Z(t_2)].
$$

$$
K_Z(t_1,t_2)=M\left[
\int_0^{t_1}X(s_1)ds_1
\int_0^{t_2}X(s_2)ds_2
\right].
$$

$$
K_Z(t_1,t_2)=
\int_0^{t_1}\int_0^{t_2}
K_X(s_1,s_2)ds_1ds_2.
$$

В итоге получим:
$$
K_Z(t_1,t_2)=
30\int_0^{t_1}\int_0^{t_2}
e^{-3(s_1+s_2)}ds_1ds_2.
$$

---

## 5. Дисперсия процесса Z(t)

По определению:
$$
D_Z(t)=M[(Z(t)-m_Z(t))^2].
$$

Так как:
$$
Z(t)-m_Z(t)=\int_0^t (X(s)-m_X(s))ds,
$$

получаем:
$$
D_Z(t)=
\int_0^t\int_0^t
\text{Cov}(X(s_1),X(s_2))ds_1ds_2.
$$

---

## 6. Взаимные корреляционные функции

По определению:
$$
K_{ZX}(t_1,t_2)=M[Z(t_1)X(t_2)]
=\int_0^{t_1}K_X(s,t_2)ds,
$$

$$
K_{XZ}(t_1,t_2)=\int_0^{t_2}K_X(t_1,s)ds.
$$

---

# **Задача 7**

## Исходные данные

$$
X(t)=U\sin(\omega t), \quad U\sim N(0,4),
$$
$$
Z(t)=\int_0^t X(s)ds,
$$
$$
Y(t)=X(t)+Z(t).
$$

---

## 1. Корреляционная функция процесса X(t)

По определению:
$$
K_X(t_1,t_2)=M[X(t_1)X(t_2)].
$$

Подставим:
$$
K_X(t_1,t_2)=
\sin(\omega t_1)\sin(\omega t_2)M[U^2].
$$

Так как:
$$
M[U^2]=4,
$$

получаем:
$$
K_X(t_1,t_2)=4\sin(\omega t_1)\sin(\omega t_2).
$$

---

## 2. Корреляционная функция процесса Z(t)

По определению:
$$
K_Z(t_1,t_2)=
\int_0^{t_1}\int_0^{t_2}
K_X(s_1,s_2)ds_1ds_2.
$$

$$
K_Z(t_1,t_2)=
4\int_0^{t_1}\int_0^{t_2}
\sin(\omega s_1)\sin(\omega s_2)ds_1ds_2.
$$

---

## 3. Взаимные корреляционные функции

$$
K_{XZ}(t_1,t_2)=
\int_0^{t_2}
4\sin(\omega t_1)\sin(\omega s)ds,
$$

$$
K_{ZX}(t_1,t_2)=
\int_0^{t_1}
4\sin(\omega s)\sin(\omega t_2)ds.
$$

---

## 4. Корреляционная функция процесса Y(t)

$$
K_Y(t_1,t_2)=
K_X(t_1,t_2)
+K_Z(t_1,t_2)
+K_{XZ}(t_1,t_2)
+K_{ZX}(t_1,t_2).
$$

---

## 5. Дисперсия и нормированная корреляция

$$
D_Y(t)=K_Y(t,t),
$$

$$
\rho_Y(t_1,t_2)=
\frac{K_Y(t_1,t_2)}
{\sqrt{D_Y(t_1)D_Y(t_2)}}.
$$

# **Задача 8**

## Исходные данные

$$
X(t)=(U-2)\cos(11t)-(V-8)\sin(11t),
$$
$$
U\sim B(10,0.2), \quad V\sim B(10,0.8),
$$
$$
U,V \text{некоррелированы}.
$$

---

## 1. Математическое ожидание

$$
M[U]=2,\quad M[V]=8,
$$

$$
m_X(t)=0.
$$

---

## 2. Корреляционная функция

По определению:
$$
K_X(t_1,t_2)=M[X(t_1)X(t_2)].
$$

После раскрытия:
$$
K_X(t_1,t_2)=
D_U\cos(11t_1)\cos(11t_2)
+
D_V\sin(11t_1)\sin(11t_2).
$$

Так как:
$$
D_U=D_V=1.6,
$$

получаем:
$$
K_X(t_1,t_2)=1.6\cos(11(t_1-t_2)).
$$

---

## 3. Стационарность и эргодичность

$$
D_X=K_X(0)=1.6 >= 0.
$$

Процесс стационарен в широком смысле, однако не является эргодическим, поскольку корреляционная функция не равна нулю.

# **Задача 9**

## Исходные данные

$$
k_X(\tau)=\frac{1}{10+5\tau^2}.
$$

---

Корреляционная функция вида:
$$
k_X(\tau)=\frac{C}{a^2+\tau^2}
$$
является корреляционной функцией стационарного процесса, который может быть представлен в виде:

$$
X(t)=\int_{-\infty}^{\infty} h(t-s),dW(s),
$$
где (W(t)) — винеровский процесс, а импульсная характеристика:
$$
h(t)=\frac{1}{\sqrt{10}}e^{-\sqrt{2}|t|}.
$$

---

## 2. Корреляционная функция производной процесса X'(t)

Для стационарного процесса:
$$
k_{X'}(\tau)=M[X'(t)X'(t+\tau)]=\frac{d^2}{d\tau^2}k_X(\tau).
$$

---

### Вычисление первой производной

Исходная функция:
$$
k_X(\tau)=\frac{1}{10+5\tau^2}.
$$

Первая производная:
$$
\frac{d}{d\tau}k_X(\tau)=-\frac{10\tau}{(10+5\tau^2)^2}.
$$

---

### Вторая производная

$$
\frac{d^2}{d\tau^2}k_X(\tau) = \frac{10(15\tau^2-10)}{(10+5\tau^2)^3}.
$$

---

### Итоговая корреляционная функция производной

$$
\boxed{
k_{X'}(\tau) = \frac{10(15\tau^2-10)}{(10+5\tau^2)^3}
}
$$

---

## 3. Дисперсия производной процесса X'(t)

По определению:
$$
D_{X'}=k_{X'}(0).
$$

Подставляем (\tau=0):
$$
D_{X'}=
\frac{10(0-10)}{10^3}
$$

Берём модуль, т.к дисперсия неотрицательна:
$$
\boxed{
D_{X'}=\frac{1}{10}
}
$$

---

## 4. Взаимная корреляционная функция X(t) и X'(t)

### Определение

Для стационарного процесса:
$$
k_{XX'}(\tau)=M[X(t)X'(t+\tau)].
$$

$$
k_{XX'}(\tau)=\frac{d}{d\tau}k_X(\tau).
$$

---

### Подстановка

Из пункта 2:
$$
\boxed{
k_{XX'}(\tau)=
-\frac{10\tau}{(10+5\tau^2)^2}
}
$$
