# Аналитический расчет характеристик случайного процесса $\zeta(t)$

## Исходные данные
Случайный процесс определяется как:
$$
\zeta(t) = \psi(t) + \chi(t),
$$
где
$$
\psi(t) = A \cos(\omega t + \varphi), \quad \chi(t) = B \cos(\Omega t),
$$
с параметрами $A = 5$, $B = 3$, и случайными величинами:
- $\varphi \sim U(-\pi, \pi)$ — равномерное распределение на интервале $(-\pi, \pi)$,
- $\Omega \sim U(\Omega_1, \Omega_2)$ — равномерное распределение на интервале $(\Omega_1, \Omega_2)$, где $\Omega_1 = 1$, $\Omega_2 = 2$.

Предполагается, что $\varphi$ и $\Omega$ независимы.

---

## 1. Математическое ожидание $m_\zeta(t)$

Математическое ожидание суммы процессов:
$$
m_\zeta(t) = M[\zeta(t)] = M[\psi(t) + \chi(t)] = M[\psi(t)] + M[\chi(t)].
$$

### Для процесса $\psi(t)$:
$$
m_\psi(t) = M[A \cos(\omega t + \varphi)] = A \cdot M[\cos(\omega t + \varphi)].
$$
Учитывая, что $\varphi \sim U(-\pi, \pi)$ с плотностью $p_\varphi(\varphi) = \frac{1}{2\pi}$:
$$
M[\cos(\omega t + \varphi)] = \int_{-\pi}^{\pi} \cos(\omega t + \varphi) \cdot \frac{1}{2\pi}  d\varphi = 0,
$$
так как интеграл косинуса по полному периоду равен нулю. Следовательно:
$$
m_\psi(t) = 0.
$$

### Для процесса $\chi(t)$:
$$
m_\chi(t) = M[B \cos(\Omega t)] = B \cdot M[\cos(\Omega t)].
$$
Учитывая, что $\Omega \sim U(1, 2)$ с плотностью $p_\Omega(\omega) = 1$:
$$
M[\cos(\Omega t)] = \int_{1}^{2} \cos(\omega t) \cdot 1  d\omega = \frac{1}{t} \left[ \sin(\omega t) \right]_{1}^{2} = \frac{\sin(2t) - \sin(t)}{t}.
$$
Таким образом:
$$
m_\chi(t) = B \cdot \frac{\sin(2t) - \sin(t)}{t} = 3 \cdot \frac{\sin(2t) - \sin(t)}{t}.
$$

### Итоговое математическое ожидание:
$$
m_\zeta(t) = m_\psi(t) + m_\chi(t) = 0 + 3 \cdot \frac{\sin(2t) - \sin(t)}{t} = \frac{3 (\sin(2t) - \sin(t))}{t}.
$$

---

## 2. Дисперсия $D_\zeta(t)$

Дисперсия суммы независимых процессов:
$$
D_\zeta(t) = D[\psi(t) + \chi(t)] = D[\psi(t)] + D[\chi(t)].
$$

### Для процесса $\psi(t)$:
$$
D_\psi(t) = M[\psi^2(t)] - (m_\psi(t))^2 = M[\psi^2(t)].
$$
Так как $m_\psi(t) = 0$, то:
$$
D_\psi(t) = M[A^2 \cos^2(\omega t + \varphi)] = A^2 \cdot M[\cos^2(\omega t + \varphi)].
$$
Используя тождество $\cos^2(x) = \frac{1 + \cos(2x)}{2}$:
$$
M[\cos^2(\omega t + \varphi)] = \int_{-\pi}^{\pi} \frac{1 + \cos(2\omega t + 2\varphi)}{2} \cdot \frac{1}{2\pi}  d\varphi = \frac{1}{2} + \frac{1}{2} M[\cos(2\omega t + 2\varphi)].
$$
Интеграл от $\cos(2\omega t + 2\varphi)$ по периоду равен нулю:
$$
M[\cos(2\omega t + 2\varphi)] = 0.
$$
Следовательно:
$$
M[\cos^2(\omega t + \varphi)] = \frac{1}{2}, \quad D_\psi(t) = A^2 \cdot \frac{1}{2} = \frac{25}{2}.
$$

### Для процесса $\chi(t)$:
$$
D_\chi(t) = M[\chi^2(t)] - (m_\chi(t))^2.
$$
Вычислим:
$$
M[\chi^2(t)] = M[B^2 \cos^2(\Omega t)] = B^2 \cdot M[\cos^2(\Omega t)].
$$
Используя тождество $\cos^2(x) = \frac{1 + \cos(2x)}{2}$:
$$
M[\cos^2(\Omega t)] = \int_{1}^{2} \frac{1 + \cos(2\omega t)}{2} \cdot 1  d\omega = \frac{1}{2} \int_{1}^{2} d\omega + \frac{1}{2} \int_{1}^{2} \cos(2\omega t)  d\omega.
$$
Первый интеграл:
$$
\int_{1}^{2} d\omega = 1.
$$
Второй интеграл:
$$
\int_{1}^{2} \cos(2\omega t)  d\omega = \frac{1}{2t} \left[ \sin(2\omega t) \right]_{1}^{2} = \frac{\sin(4t) - \sin(2t)}{2t}.
$$
Таким образом:
$$
M[\cos^2(\Omega t)] = \frac{1}{2} + \frac{\sin(4t) - \sin(2t)}{4t},
$$
$$
M[\chi^2(t)] = B^2 \left( \frac{1}{2} + \frac{\sin(4t) - \sin(2t)}{4t} \right) = 9 \left( \frac{1}{2} + \frac{\sin(4t) - \sin(2t)}{4t} \right).
$$
Теперь:
$$
D_\chi(t) = 9 \left( \frac{1}{2} + \frac{\sin(4t) - \sin(2t)}{4t} \right) - \left( \frac{3 (\sin(2t) - \sin(t))}{t} \right)^2.
$$

### Итоговая дисперсия:
$$
D_\zeta(t) = D_\psi(t) + D_\chi(t) = \frac{25}{2} + 9 \left( \frac{1}{2} + \frac{\sin(4t) - \sin(2t)}{4t} \right) - \frac{9 (\sin(2t) - \sin(t))^2}{t^2}.
$$

---

## 3. Ковариационная функция $K_\zeta(t_1, t_2)$

Ковариационная функция суммы независимых процессов:
$$
K_\zeta(t_1, t_2) = K_\psi(t_1, t_2) + K_\chi(t_1, t_2).
$$

### Для процесса $\psi(t)$:
$$
K_\psi(t_1, t_2) = M[\psi(t_1) \psi(t_2)] - m_\psi(t_1) m_\psi(t_2) = M[\psi(t_1) \psi(t_2)].
$$
Подставляем:
$$
M[\psi(t_1) \psi(t_2)] = M[A^2 \cos(\omega t_1 + \varphi) \cos(\omega t_2 + \varphi)].
$$
Используем тождество:
$$
\cos(\alpha) \cos(\beta) = \frac{1}{2} \left( \cos(\alpha - \beta) + \cos(\alpha + \beta) \right).
$$
Тогда:
$$
M[\psi(t_1) \psi(t_2)] = \frac{A^2}{2} M[\cos(\omega (t_1 - t_2)) + \cos(\omega (t_1 + t_2) + 2\varphi)].
$$
Первое слагаемое не зависит от $\varphi$, второе:
$$
M[\cos(\omega (t_1 + t_2) + 2\varphi)] = \int_{-\pi}^{\pi} \cos(\omega (t_1 + t_2) + 2\varphi) \cdot \frac{1}{2\pi}  d\varphi = 0.
$$
Следовательно:
$$
K_\psi(t_1, t_2) = \frac{A^2}{2} \cos(\omega (t_1 - t_2)) = \frac{25}{2} \cos(\omega (t_1 - t_2)).
$$

### Для процесса $\chi(t)$:
$$
K_\chi(t_1, t_2) = M[\chi(t_1) \chi(t_2)] - m_\chi(t_1) m_\chi(t_2).
$$
Вычислим:
$$
M[\chi(t_1) \chi(t_2)] = M[B^2 \cos(\Omega t_1) \cos(\Omega t_2)] = B^2 M[\cos(\Omega t_1) \cos(\Omega t_2)].
$$
Используем тождество:
$$
\cos(\alpha) \cos(\beta) = \frac{1}{2} \left( \cos(\alpha - \beta) + \cos(\alpha + \beta) \right).
$$
Тогда:
$$
M[\chi(t_1) \chi(t_2)] = \frac{B^2}{2} M[\cos(\Omega (t_1 - t_2)) + \cos(\Omega (t_1 + t_2))].
$$
Учитывая распределение $\Omega$:
$$
M[\cos(\Omega \tau)] = \int_{1}^{2} \cos(\omega \tau) \cdot 1  d\omega = \frac{\sin(2\tau) - \sin(\tau)}{\tau}.
$$
Таким образом:
$$
M[\chi(t_1) \chi(t_2)] = \frac{9}{2} \left( \frac{\sin(2(t_1 - t_2)) - \sin(t_1 - t_2)}{t_1 - t_2} + \frac{\sin(2(t_1 + t_2)) - \sin(t_1 + t_2)}{t_1 + t_2} \right).
$$
Теперь:
$$
K_\chi(t_1, t_2) = \frac{9}{2} \left( \frac{\sin(2(t_1 - t_2)) - \sin(t_1 - t_2)}{t_1 - t_2} + \frac{\sin(2(t_1 + t_2)) - \sin(t_1 + t_2)}{t_1 + t_2} \right) - m_\chi(t_1) m_\chi(t_2),
$$
где
$$
m_\chi(t_1) m_\chi(t_2) = \frac{9 (\sin(2t_1) - \sin(t_1)) (\sin(2t_2) - \sin(t_2))}{t_1 t_2}.
$$

### Итоговая ковариационная функция:
$$
K_\zeta(t_1, t_2) = \frac{25}{2} \cos(\omega (t_1 - t_2)) + \frac{9}{2} \left( \frac{\sin(2(t_1 - t_2)) - \sin(t_1 - t_2)}{t_1 - t_2} + \frac{\sin(2(t_1 + t_2)) - \sin(t_1 + t_2)}{t_1 + t_2} \right) - \frac{9 (\sin(2t_1) - \sin(t_1)) (\sin(2t_2) - \sin(t_2))}{t_1 t_2}.
$$

## Задание 2. Статистическое моделирование

### 1. Задать числовые значения для всех неслучайных параметров.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
# @title Визуализация случайного гармонического колебания
def get_random_process_realization(
    t: np.linspace,
    num_trajectories: int = 1000
) -> np.array:
    xi_t_values = []
    plt.figure(figsize=(8, 4.8))
    for _ in range(num_trajectories):
        phi = np.random.uniform(-np.pi, np.pi)  # Случайная фаза процесса psi_t
        omega_psi = 1  # Фиксированная частота
        psi_t = 5 * np.cos(omega_psi * t + phi)
        omega_chi = np.random.uniform(1, 2)  # Случайная фаза процесса chi_t
        chi_t = 3 * np.cos(omega_chi * t)
        xi_t = psi_t + chi_t
        xi_t_values.append(xi_t)
        plt.plot(t, xi_t)
    plt.xlabel('Время (t)')
    plt.ylabel('$\\xi(t)$')
    plt.title('Случайные гармонические колебания $\\xi(t) = 5\\cdot\\cos(\\omega t + \\phi) + 3\\cdot\\cos(\\Omega t)$, $Omega\\sim U[1, 2]$, $\\phi\\sim U[-\\pi, \\pi]$')
    plt.grid(True, linestyle='--')
    plt.show()
    return xi_t_values
```

```python
end_time=10
num_of_dots=100
num_trajectories=1000
```

```python
t = np.linspace(0, end_time, num_of_dots)
```

```python
xi_t_values = get_random_process_realization(
    t=t,
    num_trajectories=1000
)
```

```python
xi_t_mean = np.mean(xi_t_values, axis=0) # массив средних значений для каждого момента времени
xi_t_mean
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel('Время (t)')
plt.ylabel('M$\\xi(t)$')
plt.plot(t, xi_t_mean)
```

```python
xi_t_variance = np.var(xi_t_values, axis=0, ddof=1)
xi_t_variance
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel('Время (t)')
plt.ylabel('D$\\xi(t)$')
plt.plot(t, xi_t_variance)
```

```python
A = 5
B = 3
# Ковариационная функция
t_grid = np.linspace(0, 10, 100)
xi_interp = np.array([
    np.interp(t_grid, t, xi_t_values[i]) for i in range(num_trajectories)
])
m_interp = np.mean(xi_interp, axis=0)

K_hat = np.zeros((len(t_grid), len(t_grid)))
for k in range(len(t_grid)):
    for l in range(len(t_grid)):
        K_hat[k, l] = np.mean((xi_interp[:, k] - m_interp[k]) * (xi_interp[:, l] - m_interp[l]))
```

```python
# Теоретические функции
def m_theoretical(t):
    return B * (np.sin(2 * t) - np.sin(t)) / t

def D_theoretical(t):
    return (A**2) / 2 + B**2 * (
        0.5 + (np.sin(4 * t) - np.sin(2 * t)) / (4 * t)
        - ((np.sin(2 * t) - np.sin(t)) / t) ** 2
    )
m_theor = m_theoretical(t)
D_theor = D_theoretical(t)
```

```python
# Визуализация
plt.figure(figsize=(12, 8))

# 10 случайных реализаций процесса
plt.subplot(2, 2, 1)
for i in range(10):
    plt.plot(t, xi_t_values[i], alpha=0.6)
plt.xlabel('Время t')
plt.ylabel('$\\xi(t)$')
plt.title('10 реализаций процесса $\\xi(t)$')
plt.grid(True, linestyle='--')

# Математическое ожидание
plt.subplot(2, 2, 2)
plt.plot(t, xi_t_mean, label='Выборочное $\\hat{m}_\\xi(t)$', color='b')
plt.plot(t, m_theor, '--', label='Теоретическое $m_\\xi(t)$', color='r', linewidth=2)
plt.xlabel('Время t')
plt.ylabel('$m_\\xi(t)$')
plt.title('Сравнение теоретического и выборочного $m_\\xi(t)$')
plt.legend()
plt.grid(True, linestyle='--')

# Дисперсия
plt.subplot(2, 2, 3)
plt.plot(t, xi_t_variance, label='Выборочная $\\hat{D}_\\xi(t)$', color='b')
plt.plot(t, D_theor, '--', label='Теоретическая $D_\\xi(t)$', color='r', linewidth=2)
plt.xlabel('Время t')
plt.ylabel('$D_\\xi(t)$')
plt.title('Сравнение теоретической и выборочной дисперсии')
plt.legend()
plt.grid(True, linestyle='--')

# Диагональный срез ковариационной функции
plt.subplot(2, 2, 4)
plt.plot(t_grid, np.diag(K_hat), 'g', linewidth=2)
plt.xlabel('$t$')
plt.ylabel('$K_\\xi(t,t)$')
plt.title('Выборочная ковариационная функция (диагональ)')
plt.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
# 3D-график ковариационной функции
T1, T2 = np.meshgrid(t_grid, t_grid)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(T1, T2, K_hat, cmap='viridis', alpha=0.9)
ax.set_xlabel('$t_1$')
ax.set_ylabel('$t_2$')
ax.set_zlabel('$\\hat{K}_\\xi(t_1,t_2)$')
ax.set_title('Выборочная ковариационная функция процесса $\\xi(t)$')
fig.colorbar(surf, ax=ax, shrink=0.6)
plt.show()
```

### Рассчитать теоретическую Kζ (t1 , t2 ) на сетке значений t1 , t2 ∈ [t0 , tf ].

```python
def K_zeta_theoretical(t1, t2):
    # Kψ
    K_psi = (A**2 / 2) * np.cos(1 * (t1 - t2))
    
    # Kχ (с контролем деления на ноль)
    delta1 = t1 - t2
    delta2 = t1 + t2
    term1 = np.zeros_like(delta1)
    term2 = np.zeros_like(delta2)
    
    # Чтобы избежать деления на 0 — подставляем предел sin(x)/x → 1
    small = 1e-8
    mask1 = np.abs(delta1) < small
    mask2 = np.abs(delta2) < small

    term1[~mask1] = (np.sin(Omega2 * delta1[~mask1]) - np.sin(Omega1 * delta1[~mask1])) / delta1[~mask1]
    term1[mask1] = Omega2 - Omega1
    
    term2[~mask2] = (np.sin(Omega2 * delta2[~mask2]) - np.sin(Omega1 * delta2[~mask2])) / delta2[~mask2]
    term2[mask2] = Omega2 - Omega1

    K_chi = (B**2 / (4 * (Omega2 - Omega1))) * (term1 + term2)
    
    return K_psi + K_chi

K_theor = K_zeta_theoretical(t1, t2)
```

```python
fig = plt.figure(figsize=(12, 5))

# Теоретическая поверхность
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(t1, t2, K_theor, cmap='viridis', alpha=0.9)
ax.set_xlabel('$t_1$')
ax.set_ylabel('$t_2$')
ax.set_zlabel('$K_\\zeta(t_1, t_2)$')
ax.set_title('Теоретическая ковариационная функция $K_\\zeta(t_1, t_2)$')

# Выборочная поверхность
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(t1, t2, K_hat, cmap='plasma', alpha=0.9)
ax2.set_xlabel('$t_1$')
ax2.set_ylabel('$t_2$')
ax2.set_zlabel('$\\hat{K}_\\zeta(t_1, t_2)$')
ax2.set_title('Выборочная ковариационная функция $\\hat{K}_\\zeta(t_1, t_2)$')

plt.tight_layout()
plt.show()
```

Выборочная оценка $\hat{K}_\zeta(t_1, t_2)$ хорошо приближает теоретическую форму $K_\zeta(t_1, t_2)$  при достаточно большом ансамбле реализаций ($N=1000$).
